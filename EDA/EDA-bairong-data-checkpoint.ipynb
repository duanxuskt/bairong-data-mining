```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['font.serif'] = ['SimHei']
import seaborn as sns
%matplotlib inline
p = sns.color_palette()
sns.set_style("darkgrid",{"font.sans-serif":['simhei', 'Arial']})
```

```python
import os
data_dir = "D:/GRADUATE/BAIRONG/bairong_project/data/"
def csv_f(file_name):
    return os.path.join(data_dir,file_name)

als = pd.read_csv(csv_f("als_1568089567.csv"))
```

```python
# 自定义转换函数

def standardize(value):
    """
    转换字符串/float等数字为整型
    """
    if not pd.isnull(value):
        #print(value)
        new_value = np.float(value)
        # 如果是float的字符串，会出现错误，
        return new_value
    else:
        return value

# als_.iloc[:,12].apply(standardize).value_counts(dropna=False)
```

```python
als_ = als.iloc[1:,:]
```

```python
als_df = pd.DataFrame(columns=als_.columns.values)
```

```python
for i in range(0,13):
    als_df.iloc[:,i] = als_.iloc[:,i]
```

```python
for i in range(13,len(als_.columns.values)):
    als_df.iloc[:,i] =  als_.iloc[:,i].apply(standardize)
```

```python
als_df
```

```python
als_df.iloc[:,15].value_counts(dropna=False).sort_index()
```

```python
import json

als_col = als_df.columns.values
stat = dict()
for i in range(13,len(als_df.columns.values)):
    json_str = json.dumps((als_df.iloc[1:,i].value_counts(dropna=False).sort_index()).to_dict())
    stat[als_col[i]] = json_str

    
##保存文件
import csv

with open("als_temp.csv","w",newline='') as csv_file:
	writer = csv.writer(csv_file)
	for key,value in stat.items():
		writer.writerow([key,value])
```
