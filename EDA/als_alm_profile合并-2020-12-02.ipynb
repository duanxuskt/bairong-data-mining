```python
import os
data_dir = "D:/GRADUATE\BAIRONG/Review"
def csv_f(file_name):
    return os.path.join(data_dir,file_name)
```

```python
als = pd.read_csv(csv_f("als_1568089567.csv"))
als.head(10)
```

```python
als_ = als.iloc[1:,:]
```

```python
als_
```

```python
# 统一好数据格式，首先进行drop处理，干掉大量缺失的数据
# 删除
empty_cus_num = list()
with open(r"D:\GRADUATE\BAIRONG\Review\empty_cus_num.txt",'r') as f:
    for line in f.readlines():
        empty_cus_num.append(line.strip('\n'))
print(len(empty_cus_num))
```

```python
als_['cus_num'] = als_['cus_num'].astype("str")
```

```python
als_keep = als_.drop(als_[als_['cus_num'].isin(empty_cus_num)].index)
```

```python
als_keep
```

```python
# 我们选取了508个维度，这508个维度的选取标准是nan值占比在5%以下，但是又不存在0值的，我们拟将这些维度填充为0.0
als_features = list()
with open(r"D:\GRADUATE\BAIRONG\Review\als_features.txt",'r') as f:
    for line in f.readlines():
        als_features.append(line.strip('\n'))
print(len(als_features))
```

```python
als_features_ = [als_keep[fe] for fe in als_features]
temp = [als_keep['cus_num']]
temp.extend(als_features_)
```

```python
als_keep_ = pd.concat(temp,axis=1)
```

```python
# excel表筛选出部分特征，进行缺失值处理，我们先全填充为nan，因为这些特征并不包含0值，可以将nan视同为0
for fe in als_features:
    als_keep_[fe].fillna(0.0,inplace=True)
```

```python
als_keep_
```

```python
alm_all_keep_r_distill = pd.read_csv(r"D:/GRADUATE\BAIRONG/Review/alm_all_keep_r_distill.csv")
```

```python
# 合并一下
```

```python
alsm_all_keep_merge = pd.merge(alm_all_keep_r_distill,als_keep_,how='inner',on=['cus_num'])
```
